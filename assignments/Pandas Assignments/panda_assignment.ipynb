```python
# Q6 – Convert timestamp column to datetime
# Your answer here:

df["timestamp"] = pd.to_datetime(df["timestamp"])
df['timestamp'].dtype
```

```python
# Q7 – Display unique countries and cities
# Your answer here:

unique_countries = df["country"].unique()
unique_countries
unique_cities = df["city"].unique()
unique_cities
```

```python
# Q8 – Filter data where AQI > 150
# Your answer here:

df_filtered = df[df["aqi"] > 150]
df_filtered
```

```python
# Q9 – Sort by PM2.5 descending
# Your answer here:

df.sort_values(by = "pm25", ascending = False).head(10)
```

```python
# Q10 – Group by country and calculate mean AQI
# Your answer here:

df.groupby('country')['aqi'].mean()
```

```python
# Q11 – Group by city and find top 5 for PM2.5
# Your answer here:

df.groupby('city')['pm25'].mean().nlargest(10)
```

```python
# Q12 – Find correlations of AQI with numeric columns
# Your answer here:

df.corr(numeric_only=True)['aqi']
```

```python
# Q13 – Create new feature columns (hour, day, month from timestamp)
# Your answer here:

df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
```

```python
# Q14 – Check duplicates (count & display)
# Your answer here:

df.duplicated().sum()
```

```python
# Q15 – Remove duplicates if needed
# Your answer here:

No duplicates so no need to remove the duplicate
```

```python
# Q16 – Handle missing values (drop or fill)
# Your answer here:

df.isna().sum()
df = df.dropna()
```

```python
# Q17 – Boolean filter: temperature > 25 and humidity < 40
# Your answer here:

filtered_data = df[(df['temperature'] > 25) & (df['humidity'] < 40)]
filtered_data
```

```python
# Q18 – Export cleaned dataset
# Your answer here:

df.to_csv("cleaned_dataset.csv", index = False)
```

```python
# Q19 – Identify the top 3 most correlated features with AQI using Pandas
# Your answer here:

aqi_corr = df.corr(numeric_only=True)['aqi'].drop('aqi').abs().head(10)
aqi_corr
```

```python
# Q20 – Create a country-level summary table using Pandas
# Your answer here:

country_summary = df.groupby("country").agg(mean_aqi=("aqi", "mean"),max_pm25=("pm25", "max"),min_pm10=("pm10", "min"),
avg_temperature=("temperature", "mean"),records_count=("country", "count")).reset_index().sort_values(by="mean_aqi", ascending=False)
country_summary
```
