```python
import numpy as np
```

```python
# Q1 – Create NumPy arrays (1D, 2D, 3D)

arr_1d = np.array([1, 2, 3, 4, 5])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_3d = np.array([[[1, 2,], [3, 4]], [[5, 6], [7, 8]]])
print(arr_1d)
print(arr_2d)
print(arr_3d)
```

```python
# Q2 – Print shape, size, and datatype of an array

print(arr_2d.shape)
print(arr_2d.size)
print(arr_2d.dtype)
```

```python
# Q3 – Perform indexing and slicing on a NumPy array

indexing_arr = arr_1d[-1]
print(indexing_arr)
sliced_arr = arr_1d[1::2]
print(sliced_arr)
```

```python

# Q4 – Perform arithmetic operations on arrays

squared_arr = arr_1d ** 2
print(squared_arr)
```

```python

# Q5 – Use NumPy built-in functions (sum, mean, max, min)

sum = np.sum(arr_1d)
mean = np.mean(arr_1d)
max = np.max(arr_1d)
min = np.min(arr_1d)

print(sum)
print(mean)
print(max)
print(min)
```

```python

# Q6 – Reshape a 1D array into 2D

arr = np.array([1,2,3,4,5,6])
print(arr)

reshaped_arr = arr.reshape(2,3)
print(reshaped_arr)
```

```python
# Q9 – Create arrays using zeros(), ones(), arange(), linspace()

zero_arr = np.zeros((2,2))
ones_arr = np.ones((2,2))
arrayy = np.arange(0,10,2)
line_space = np.linspace(1,10,5)
print(zero_arr)
print(ones_arr)
print(arrayy)
print(line_space)
```
