```python
# Q2 – Print shape, size, and datatype of an array

print(arr_2d.shape)
print(arr_2d.size)
print(arr_2d.dtype)
```

```python
# Q3 – Perform indexing and slicing on a NumPy array

indexing_arr = arr_1d[-1]
print(indexing_arr)
sliced_arr = arr_1d[1::2]
print(sliced_arr)
```
